### Import libraries

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv('NYC_Property_Data.csv') #import data
```

```python
data
```

### Clean data

```python
data.isnull().sum() 
```

```python
data.dropna(how ='any', inplace=True)
```

```python
data.isnull().sum()
```

### Drop unnecessary info

```python
info = data.drop(['Unnamed: 0', 'Unnamed: 0.1','SALE PRICE','ADDRESS','SALE DATE'], axis = 1)
```

```python
info.head()
```

```python
target = data['SALE PRICE']
```

### Preprocess data

```python
le = LabelEncoder()
```

```python
info['NEIGHBORHOOD'] = le.fit_transform(info['NEIGHBORHOOD'])
info['BUILDING CLASS CATEGORY'] = le.fit_transform(info['BUILDING CLASS CATEGORY'])
info['TAX CLASS AT PRESENT'] = le.fit_transform(info['TAX CLASS AT PRESENT'])
info['BUILDING CLASS AT PRESENT'] = le.fit_transform(info['BUILDING CLASS AT PRESENT'])
info['BUILDING CLASS AT TIME OF SALE'] = le.fit_transform(info['BUILDING CLASS AT TIME OF SALE'])
```

```python
info
```

### Model and Evaluate

```python
info_train, info_test, target_train, target_test = train_test_split(info, target, test_size = 0.25)
```

```python
info_train.head()
```

```python
regressor = LinearRegression()
```

```python
model = regressor.fit(info_train, target_train) #.fit goes line by line to train and test
```

```python
target_pred = model.predict(info_test)
```

### Create a frame to identify Actual vs Predicted and the Error

```python
pd.DataFrame({'Actual': target_test, 'Predicted': target_pred, 'Error': target_test-target_pred})
```
