## Removal of Unwanted Data

```python
detail = fromfile.drop(['Unnamed: 0', 'Unnamed: 0.1','SALE PRICE','ADDRESS','SALE DATE'], axis = 1)
detail.head()
```

```python
select = fromfile['SALE PRICE']
```

## Data Processing

```python
le = LabelEncoder()
detail
```

```python
detail['NEIGHBORHOOD'] = le.fit_transform(detail['NEIGHBORHOOD'])
detail['BUILDING CLASS CATEGORY'] = le.fit_transform(detail['BUILDING CLASS CATEGORY'])
detail['TAX CLASS AT PRESENT'] = le.fit_transform(detail['TAX CLASS AT PRESENT'])
detail['BUILDING CLASS AT PRESENT'] = le.fit_transform(detail['BUILDING CLASS AT PRESENT'])
detail['BUILDING CLASS AT TIME OF SALE'] = le.fit_transform(detail['BUILDING CLASS AT TIME OF SALE'])
```

```python
detail
```

## The Model

```python
detail_train, detail_test, select_train, select_test = train_test_split(detail, select, test_size = 0.25)
```

```python
detail_train.head()
```

```python
regressor = LinearRegression()
```

```python
model = regressor.fit(detail_train, select_train)
```

```python
select_pred = model.predict(detail_test)
```

## Using a Table to Show Actual vs Prediction and their Errors

```python
pd.DataFrame({'Actual': select_test, 'Predicted': select_pred, 'Error': select_test-select_pred})
```
